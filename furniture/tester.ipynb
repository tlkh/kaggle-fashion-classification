```python
from os import listdir
from os.path import isfile, join
test_set = [f for f in listdir("test2") if isfile(join("test2", f))]
print("Total:", len(test_set), "images")
print("Example:", test_set[0])
print("Missing images:", 12800-len(test_set), "(",round((12800-len(test_set))/12800, 4)*100,"% )")
```

```python
from keras.applications.xception import Xception, preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.models import Model, load_model
from keras.layers import *
from keras import backend as K
from keras import optimizers, callbacks
import numpy as np
import pandas as pd
import cv2, h5py
from random import randint
```

```python
model = load_model("xception-0.721.h5")
```

```python
from glob import glob
class_names = glob("train_aug/*") # Reads all the folders in which images are present
class_names = sorted(class_names) # Sorting them
fixed_classes = []
for class_name in class_names:
    fixed_classes.append(class_name[10:])
name_id_map = dict(zip(range(len(class_names)), fixed_classes))
```

```python
output_id, output_label = [], []
for i in range(1,12801):
    imagepath = (str(i)+".jpg")
    if imagepath in test_set:
        try:
            img = cv2.imread("test2/"+imagepath)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (229, 229)).astype(np.float32)
            img = img.reshape((1,) + img.shape)
            img = preprocess_input(img)
            preds = model.predict(img)
            output_pred = name_id_map[np.argmax(preds,axis=1)[0]]
            print(int(imagepath[:-4]), int(output_pred))
            output_id.append(int(imagepath[:-4]))
            output_label.append(int(output_pred))
        except Exception as e:
            print(e)
            output_id.append(int(imagepath[:-4]))
            output_label.append(int(randint(1, 129)))
    else:
        print("missing, rng LOL")
        output_id.append(int(imagepath[:-4]))
        output_label.append(int(randint(1, 129)))
```

```python
len(output_id), len(output_label)
```

```python
with open('results.csv', 'w') as f:
    f.write("id,predicted\n")
    for n, img in enumerate(output_id):
        f.write(str(img)+","+str(output_label[n])+"\n")
```
